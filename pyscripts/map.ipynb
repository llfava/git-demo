```python
import folium
import json
from IPython.display import HTML
import re
```

```python
def inline_map(map):
    """
    Embeds the HTML source of the map directly into the IPython notebook.
    
    This method will not work if the map depends on any files (json data). Also this uses
    the HTML5 srcdoc attribute, which may not be supported in all browsers.
    """
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))
```

```python
start_address = [37.8001497, -122.4107165]

json_bars = '../processed/bars_rated_prelim.json'
bars = json.loads(open(json_bars).read())    
```

```python
map = folium.Map(location=start_address, zoom_start=14, max_zoom=17, tiles='Stamen Terrain')
map.simple_marker(start_address, popup='Starting address', marker_icon='star')

if True:
    counter = 0
    search = ", San Francisco, CA"
    for bar in bars:
        if counter < 20:
            s = re.search("(.*)" + search, bars[bar]['bar_address'])
            address = s.group(1)   
            popup_name = '<a href="http://www.yelp.com' + bar.encode("utf8") + '">' + bars[bar]['bar_name'].encode("utf8") + "</a>"
            popup_details = address.encode("utf8") + '<br/>' + 'Yelp rating: ' + str(bars[bar]['bar_rating']).encode("utf8")
            popup_text = popup_name + '<br/>' + popup_details
            #popup_text = '<a href="http://www.yelp.com' + bar.encode("utf8") + '">' + bars[bar]['bar_name'].encode("utf8") + "</a>" + '<br/> ' + address.encode("utf8")
        
            if bars[bar]["local_rating"] == "local": 
                map.simple_marker([bars[bar]['lat'],bars[bar]['lon']], popup=popup_text, marker_color='green', marker_icon='heart')
            if bars[bar]["local_rating"] == "touristy":
                map.simple_marker([bars[bar]['lat'],bars[bar]['lon']], popup=popup_text, marker_color='red', marker_icon='camera')
            if bars[bar]["local_rating"] == "neutral":
                map.simple_marker([bars[bar]['lat'],bars[bar]['lon']], popup=popup_text, marker_color='orange', marker_icon='ok')
            counter += 1
#inline_map(map)
map.create_map(path='/Users/ldaniel/git-demo/front_end/app/templates/map_test.html')
```
