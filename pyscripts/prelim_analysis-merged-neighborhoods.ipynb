```python
import sys
import json
import numpy
import pylab
import matplotlib.pyplot as plt
from scipy import stats
```

```python
json_bars = '../processed/bars_latlon.json'

bar_dict = json.loads(open(json_bars).read())
bar_reviews = {}
for key in bar_dict:
    fname = bar_dict[key]['file_name']
    json_reviews = '../processed/' + fname + '.json'
    try:
        bar_reviews[fname] = json.loads(open(json_reviews).read())
    except IOError:
        sys.stdout.write('File not found. Skipping %s\n' % fname)
```

```python
#Check number of user reviews matches number of reviews for bar
complete_reviews = set()
bars = bar_reviews.keys()
for bar in bars:
    if bar_dict['/biz/' + bar]['bar_num_reviews'] <= len(bar_reviews[bar])+2:
        complete_reviews.add(bar)
    else:
        print bar_dict['/biz/' + bar]['bar_name']
        print bar_dict['/biz/' + bar]['bar_num_reviews'] - len(bar_reviews[bar])
num_bars_complete= len(complete_reviews)
print num_bars_complete
```

```python
#Look at number of SF reviewers to number of nonSF reviewers
location_ratios = {}
count = 0
for bar in complete_reviews:
    users = bar_reviews[bar].keys()
    if len(users) < 10:
        continue
    count += 1
    local = 0
    for user in users:
        if bar_reviews[bar][user]['user_location'] == 'San Francisco, CA':
            local += 1
    ratio = float(local) / (len(users)-local)
    location_ratios[bar] = [ratio, local, len(users) - local]
    print "%.2f, %d %d %s" % (location_ratios[bar][0], location_ratios[bar][1], location_ratios[bar][2], bar)
print count
```

```python
#Look at average local rating vs ave nonlocal rating for each bar
ave_local_rating = {}
ave_tourist_rating = {}
count = 0
for bar in complete_reviews:
    users = bar_reviews[bar].keys()
    if len(users) < 10:
        continue
    count += 1
    local = 0
    local_rating = 0
    tourist_rating = 0
    for user in users:
        if bar_reviews[bar][user]['user_location'] == 'San Francisco, CA':
            local += 1
            local_rating += bar_reviews[bar][user]['user_rating']
        else:
            tourist_rating += bar_reviews[bar][user]['user_rating']
    ave_local_rating[bar] =  float(local_rating)/local
    ave_tourist_rating[bar] = float(tourist_rating)/(len(users)-local)
                
    print "%.2f, %.2f %s" % (ave_local_rating[bar], ave_tourist_rating[bar], bar)
print count
```

```python
#Look at average local rating vs ave nonlocal rating for each bar (weighted)

ave_local_rating = {}
ave_tourist_rating = {}
count = 0
for bar in complete_reviews:
    users = bar_reviews[bar].keys()
    if len(users) < 10:
        continue
    count += 1
    local = 0
    local_rating = 0
    tourist_rating = 0
    
    for user in users:
        if bar_reviews[bar][user]['user_location'] == 'San Francisco, CA':
            local += 1
            local_rating += bar_reviews[bar][user]['user_rating']
        else:
            tourist_rating += bar_reviews[bar][user]['user_rating']
    ave_local_rating[bar] =  float(local_rating)/len(users)
    ave_tourist_rating[bar] = float(tourist_rating)/len(users)
    
    print "%.2f, %.2f, %s" % (ave_local_rating[bar], ave_tourist_rating[bar], bar)
print count
```

```python

x_list = []
y_list = []
for bar in ave_local_rating:
    x_list.append(ave_local_rating[bar])
    y_list.append(ave_tourist_rating[bar])
x_median = (numpy.median(x_list))
y_median = (numpy.median(y_list))

rated_bars = {}

if True:
    for bar in ave_local_rating:
        rated_bars['/biz/' + bar] = bar_dict['/biz/' + bar]
        if ave_local_rating[bar] > x_median and ave_tourist_rating[bar] < y_median:
            rated_bars['/biz/' + bar]['local_rating'] = 'local'
        elif ave_local_rating[bar] < x_median and ave_tourist_rating[bar] > y_median:
            rated_bars['/biz/' + bar]['local_rating'] = 'touristy'
        else:
            rated_bars['/biz/' + bar]['local_rating'] = 'neutral'
                
            
if False:
    with open('../processed/bars_rated_prelim.json', 'w') as outfile:
        json.dump(rated_bars, outfile, indent=2)
```

```python
#visualize ave sf rating vs ave nonsf rating for each bar
x_list = []
y_list = []
for bar in ave_local_rating:
    x_list.append(ave_local_rating[bar])
    y_list.append(ave_tourist_rating[bar])
print numpy.median(x_list)
print numpy.median(y_list)
    
x = numpy.array(x_list)
y = numpy.array(y_list)

correlation = stats.pearsonr(x,y)
corr = 'r = ' + str(round(correlation[0],2))

fig = plt.figure()
ax1 = plt.subplot((111), aspect = 1)
#ax1.text(3.0, 3.5, corr, fontsize=15)
ax1.plot(x,y, 'bo')


ax1.set_xlabel('average local score', fontsize=15)
ax1.set_ylabel('average tourist score', fontsize=15 )


v = [0, 4, 0, 4]
ax1.axis(v)
ax1.plot()
#plt.show()

pylab.savefig('ave_rating.png')
```
